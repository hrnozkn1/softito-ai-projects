## Analiz ve Sonuçlar

### Duygu Analizi (VADER)

Her film yorumu için VADER kullanılarak bir `compound` (bileşik) duygu skoru hesaplanmıştır. Bu skor -1 (en negatif) ile +1 (en pozitif) arasında değişmektedir. Belirlenen eşik değerlerine göre (`compound >= 0.05` için pozitif, `compound <= -0.05` için negatif, aksi takdirde nötr) VADER tahminleri (`vader_prediction`) oluşturulmuştur.

### Temel Bulgular

*   **Doğruluk Oranı:** VADER algoritmasının bu veri seti üzerindeki doğruluk oranı yaklaşık **%69.33** olarak bulunmuştur.
*   **Skor Dağılımı:** VADER `compound` skorlarının dağılımı görselleştirilmiş, çoğu yorumun aşırı negatif veya aşırı pozitif skorlara sahip olduğu gözlemlenmiştir.
*   **Hatalı Tahmin Analizi:** Modelin yanlış tahmin ettiği yorumlar detaylı olarak incelenmiştir. Bu analizde, VADER'ın özellikle **sarkazm, ironi, ince mizah** veya **karmaşık cümle yapıları** içeren yorumlarda zorlandığı tespit edilmiştir. Örneğin, yorumun başında pozitif kelimeler kullanılıp devamında eleştirel bir dönüş yapıldığında veya tam tersi durumlarda modelin yanıldığı görülmüştür.

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk

# 1. VADER Sözlüğünü ve Görsel Tema Ayarlarını Yapıyoruz
nltk.download('vader_lexicon')
sns.set_theme(style="whitegrid") # Grafiklerimizin şık görünmesi için

# 2. Veri Setini Yüklüyoruz
# NOT: Bilgisayarına inen CSV dosyasının adı neyse buraya onu yazmalısın (Örn: "train.csv")
df = pd.read_csv("Train.csv")

# *** OPSİYONEL: Veriyi küçültmek isterseniz aşağıdaki satırların yorumunu kaldırın ***
# df = df.sample(n=10000, random_state=42) # Rastgele 10.000 satır seç
# df = df.head(10000) # İlk 10.000 satırı seç
# print(f"\nVeri seti {len(df)} satıra küçültüldü.")

# İlk 5 satıra bakarak sütun isimlerini kontrol edelim (Genelde 'text' ve 'label' olur)
print("--- Verinin İlk 5 Satırı ---")
print(df.head())

# 3. VADER ile Skorlama Yapıyoruz (Döngüsüz, Hızlı Yöntem)
sia = SentimentIntensityAnalyzer()

print("\nVeriler analiz ediliyor, lütfen bekleyin...")
# 'text' sütunundaki her yorum için compound (genel duygu) skorunu hesaplıyoruz
df['compound'] = df['text'].apply(lambda x: sia.polarity_scores(str(x))['compound'])

# 4. Skora Göre Kendi Tahminimizi Yapıyoruz
def skor_etiketle(skor):
    if skor >= 0.05:
        return 1  # Pozitif
    elif skor <= -0.05:
        return 0  # Negatif
    else:
        return -1 # Nötr (IMDB'de genelde 0 ve 1 vardır ama VADER nötr de bulabilir)

df['vader_prediction'] = df['compound'].apply(skor_etiketle)

# 5. GRAFİK ZAMANI: Sonuçları Görselleştirelim
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Grafik 1: VADER Skorlarının Dağılımı (Yoğunluk Grafiği)
sns.histplot(data=df, x='compound', kde=True, ax=axes[0], color='teal', bins=30)
axes[0].set_title('IMDB Yorumlarının VADER Compound Skor Dağılımı')
axes[0].set_xlabel('Duygu Skoru (-1: Çok Negatif, +1: Çok Pozitif)')
axes[0].set_ylabel('Yorum Sayısı')

# Grafik 2: Gerçek Etiketler ile VADER Tahminlerinin Karşılaştırılması
# Veri setindeki orijinal sütun adının 'label' olduğunu varsayıyoruz (0=neg, 1=pos)
sns.countplot(data=df, x='label', hue='vader_prediction', ax=axes[1], palette='Set2')
axes[1].set_title('Gerçek Sınıflar vs VADER Tahminleri')
axes[1].set_xlabel('Gerçek Sınıf (0: Negatif, 1: Pozitif)')
axes[1].set_ylabel('Yorum Sayısı')
axes[1].legend(title='VADER Tahmini', labels=['Nötr (-1)', 'Negatif (0)', 'Pozitif (1)'])

plt.tight_layout()
plt.show()
df.tail()

# 6. Başarı Analizi (Accuracy)
# Nötrleri hesaba katmadan VADER'ın gerçek etiketleri ne oranda bildiğine bakalım
dogru_tahminler = df[df['label'] == df['vader_prediction']]
accuracy = len(dogru_tahminler) / len(df) * 100
print(f"\nVADER Algoritmasının Bu Veri Setindeki Doğruluk Payı (Accuracy): %{accuracy:.2f}")
````

````python
print('\n--- İlk 10 Yorum ve VADER Tahminleri ---')
display(df[['text', 'label', 'vader_prediction']].head(10))
````

````python
# VADER'ın yanlış tahmin ettiği yorumları filtrele
incorrect_predictions = df[df['label'] != df['vader_prediction']]

print('\n--- VADER Tarafından Yanlış Tahmin Edilen İlk 5 Yorum ---')
display(incorrect_predictions[['text', 'label', 'vader_prediction']].head(5))
````

````python
# Yanlış tahmin edilen yorumları compound skoruna göre sırala (en negatiften en pozitife)
incorrect_predictions_sorted = incorrect_predictions.sort_values(by='compound', ascending=True)

print('\n--- En Negatif Compound Skoruna Sahip Yanlış Tahmin Edilen İlk 5 Yorum ---')
display(incorrect_predictions_sorted[['text', 'label', 'compound', 'vader_prediction']].head(5))

print('\n--- En Pozitif Compound Skoruna Sahip Yanlış Tahmin Edilen İlk 5 Yorum ---')
display(incorrect_predictions_sorted[['text', 'label', 'compound', 'vader_prediction']].tail(5))
````
